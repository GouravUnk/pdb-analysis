```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
# Load your JSON data
with open("../data/europepmc_articles.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Create DataFrame
df = pd.DataFrame(data)
```

```python
# Flatten all grant entries into one list using pandas
all_grants_series = df['grants'].explode()
grant_counts = all_grants_series.value_counts()
grant_counts
```

```python
grant_counts.head(20).sort_values().plot(
    kind='barh',
    figsize=(12, 20),
    title='All Grant Agencies by Frequency',
    xlabel='Number of Grants',
    ylabel='Grant Agency',
    color='skyblue'
)
plt.show()
```
